# Compiler Explorer With CFiddle


[Compiler Explorer](https://godbolt.org/) is a great tool for visualizing how compilers convert code to assembly, and it served as an import inspiration for CFiddle. 

Here's a version of compiler explorer written using CFiddle.  It works best in Jupyter Lab where you can open [compiler_explorer/explore.cpp](compiler_explorer/explore.cpp) in a separate editor window like this:


Then you can run the cell below, edit your code.  Each time you save the CFG will update.  The only bit of "magic" here is `changes_in()` which watches the source file for changes.

```python
%xmode Minimal
from cfiddle import *
from IPython.display import clear_output

for i in changes_in("compiler_explorer/explore.cpp"):
    clear_output(wait=True)
    c = build("compiler_explorer/explore.cpp", build_parameters=arg_map(OPTIMIZE="-O1"))
    display(c[0].cfg("go"))
```
